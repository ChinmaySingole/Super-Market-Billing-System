```python
Products = {
    "Apple": 40,
    "Banana": 10,
    "Chips": 20,
    "Avocado": 105,
    "Cola": 40,
    "Milk": 50,
    "Bread": 30,
    "Rice": 60,
    "Sugar": 45,
    "Salt ": 20,
    "Butter": 90,
    "Cheese": 120,
    "Chicken": 250,
    "Eggs ": 70,
    "Shampoo": 150,
    "Soap": 40,
    "Toothpaste": 55,
    "Detergent": 80,
    "Cooking Oil": 200,
    "Tea": 110,
    "Coffee": 130,
    "Tomato": 40,
    "Potato": 60,
    "Onion": 30
}
```

```python

print("-"*15, "Welcome To Super Market","-"*15)
while True:
    cart ={}
    amount = 0
    name = input('Enter Your Name: ')
    ph_no = input( "Enter Phone Number: ")
    while len (ph_no) != 10 or not ph_no.isdigit():
        print("Please enter a valid 10-digit phone number.")
        phone_number = input("Enter your phone number: ")
    discount = 0
    
    print("-" * 50)
    print("{:<20}{:<10}".format("Product Name", "Price"))
    print("-" * 50)

    for i, j in Products.items():
        print("{:<20}{:<10}".format(i, j))
    
    while True:
        p_name = input("Enter Product Name You Wish To Buy : ").capitalize().strip()
        quantity = float(input("Enter Quantity You Wish to Buy: "))
        val = []
        val.append(Products[p_name])
        val.append (quantity)
        val.append(quantity*Products[p_name] )
        cart[p_name] = val
        amount+= Products[p_name]*quantity
        repeat = input("Do You want to Add More Items? Yes/No : ").capitalize()
        if repeat == "No":
            break
            
    if amount <= 500:
        discount = 0
    elif amount <= 1000:
        discount = 0.05
    elif amount <= 3000:
        discount = 0.10
    else:
        discount = 0.15
        
    act_amount = amount - (amount *discount)
    print ("-"*50)
    print("-"*22,"BILL","-"*22)
    print("Customer Name: ", name)
    print("Phone Number: ",ph_no)
    print("{:<20}{:<10}{:<10}{:<10}".format("Description", "Rate", "Qty", "Amount"))
  
    for i, j in cart.items():
        print("{:<20}{:<10}{:<10}{:<10}".format(i, j[0], j[1], j[2]))
      
    print("Total Amount: ", amount)
    print( "Discount Applied : ", discount*100, "%")
    print ("-"*50)
    print("Saved on MRP : ",amount - act_amount)
    print("Payable Amount : ", act_amount)    
    print("-"*50)
    print("Thank you for shopping with Supermarket!") 
    break
```
